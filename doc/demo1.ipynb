## Analysis & Visualization: Time plotting using BeakerX plot libraries

```python
ch = Crosshair(color=Color(255, 128, 5), width=2, style=StrokeType.DOT)

plot = TimePlot(timeZone="America/New_York",
                title="Historical prices for Dollar Industries Limited",
                xLabel="Date",
                yLabel="Price",
                initWidth=950,
                initHeight=500,
                crosshair=ch)

datetime = a['Close'].index.tolist()
xs = [dt.timestamp() * 1000 for dt in datetime]
y1s = a['Open'].values.tolist()
y2s = a['Close'].values.tolist()
y3s = a['High'].values.tolist()
y4s = a['Low'].values.tolist()

plot.add(Line(x=xs, y=y1s, size=3, displayName="Open"))
plot.add(Line(x=xs, y=y2s, size=3, displayName="Close"))
plot.add(Line(x=xs, y=y3s, size=3, displayName="High"))
plot.add(Line(x=xs, y=y4s, size=3, displayName="Low"))

plot.add(ConstantLine(y=1750, style=StrokeType.DASHDOT, color=Color.blue))
plot.add(ConstantLine(x=1497250000000, style=StrokeType.DOT, color=Color.gray))

plot
```
